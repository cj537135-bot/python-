### One-sample t-test: RAM > 4 GB

```python
t_stat, p_value = stats.ttest_1samp(df['RAM'], 16, alternative='greater')
print("One-sample t-test: Mean RAM > 16 GB")
print("t-statistic:", t_stat, "p-value:", p_value, "\n")
```

#### Interpretation of Results
Since the p-value (1.0) is much greater than 0.05, the null hypothesis cannot be rejected.
Conclusion:
There is no statistical evidence to support the claim that the mean RAM is greater than 16 GB. In fact, the data strongly indicates that the average RAM is significantly lower than 16 GB.

### Two-sample t-test: Price difference between Android and iOS

There is a statistically significant difference in prices between Android and iOS smartphones, with iOS devices being more expensive on average.

# ANOVA Tests

### Price difference across brands

```python
brands_price = [df[df['brand']==b]['price(INR)'] for b in df['brand'].unique()]
f_stat, p_value = stats.f_oneway(*brands_price)
print("ANOVA: Price difference across brands")
print("F-statistic:", f_stat, "p-value:", p_value, "\n")

```

#### Interpretation of Results
Since the p-value is far less than 0.05, the null hypothesis is rejected.
Conclusion:
There is strong statistical evidence that average smartphone prices differ significantly across brands.

# Chi-Square Tests

### OS choice vs gender

```python
os_gender_table = pd.crosstab(df['operating system'], df['gender'])
chi2, p, dof, expected = stats.chi2_contingency(os_gender_table)
print("Chi-square: OS choice vs Gender")
print("Chi2:", chi2, "p-value:", p, "\n")
```

#### Interpretation of Results
Since the p-value (0.9773) is much greater than 0.05, the null hypothesis cannot be rejected.
Conclusion:
There is no statistically significant association between gender and operating system choice.

### Brand choice vs OS

```python
brand_os_table = pd.crosstab(df['brand'], df['operating system'])
chi2, p, dof, expected = stats.chi2_contingency(brand_os_table)
print("Chi-square: Brand vs OS")
print("Chi2:", chi2, "p-value:", p, "\n")
```

#### Interpretation of Results
Since the p-value is far smaller than 0.05, the null hypothesis of independence is rejected.
Conclusion:
There is a statistically significant association between smartphone brand and operating system.

### Brand choice vs occupation

```python
brand_occ_table = pd.crosstab(df['brand'], df['occupation'])
chi2, p, dof, expected = stats.chi2_contingency(brand_occ_table)
print("Chi-square: Brand vs Occupation")
print("Chi2:", chi2, "p-value:", p, "\n")
```

#### Interpretation of Results
